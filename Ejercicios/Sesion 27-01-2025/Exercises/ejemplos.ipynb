## Utilizando digits

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, NMF
from sklearn.datasets import load_digits
```

```python
# Cargar el dataset de dígitos
digits = load_digits()
X = digits.data  # Datos de las imágenes (64 características por imagen)
y = digits.target  # Etiquetas de las imágenes
```

```python
# Función para mostrar una imagen
def ShowMe(X, index=0):
    plt.imshow(X[index].reshape(8, 8), cmap='gray')
    plt.title(f"Dígito: {y[index]}")
    plt.axis('off')
    plt.show()
```

```python
# Mostrar una imagen original
ShowMe(X)
```

```python
# Función para reducir la dimensionalidad usando PCA
def reducedInputPCA(X, nPC):
    pca = PCA(n_components=nPC)
    X_pca = pca.fit_transform(X)
    return pca.inverse_transform(X_pca)
```

```python
# Visualizar imágenes reducidas con diferentes números de componentes principales
for nPC in [30, 15, 10, 5, 2]:
    Z = reducedInputPCA(X, nPC)
    print(f"Número de componentes principales (PCA) = {nPC}")
    ShowMe(Z)
```

```python
# Reducción de dimensionalidad usando NMF
def reducedInputNMF(X, n_components):
    nmf = NMF(n_components=n_components, init='random', random_state=42)
    X_nmf = nmf.fit_transform(X)
    return np.dot(X_nmf, nmf.components_)
```

```python
# Visualizar imágenes reducidas con NMF
for n_components in [30, 15, 10, 5, 2]:
    Z = reducedInputNMF(X, n_components)
    print(f"Número de componentes de NMF = {n_components}")
    ShowMe(Z)
```

```python
# Comparación de la varianza explicada por PCA
pca = PCA(n_components=64)
pca.fit(X)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
plt.plot(range(1, 65), cumulative_variance, marker='o', linestyle='--')
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulativa')
plt.title('Varianza Explicada por PCA')
plt.grid()
plt.show()
```
