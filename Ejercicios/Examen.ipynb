```python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=[name.strip().replace(" ", "_").replace("(", "").replace(")", "") for name in data.feature_names])
df = df.assign(target=pd.Series(1 - data.target))

df
```

```python
corr_matrix = df.corr()

corr = corr_matrix.round(3)

target_corr = corr['target']

target_corr_excl_last = target_corr.iloc[:-1]

target_corr_excl_last
```

```python
max_corr_index = target_corr_excl_last.idxmax()
max_corr_index
```

```python
max_corr_value = target_corr_excl_last[max_corr_index]
max_corr_value
```

```python
print(f"La característica más correlacionada con 'target' es: {max_corr_index}")
print(f"El valor de correlación más alto es: {max_corr_value}")
```
