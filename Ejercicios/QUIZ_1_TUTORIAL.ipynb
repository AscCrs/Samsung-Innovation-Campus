```python
import pandas as pd
import numpy as np
import os

from google.colab import files
uploaded = files.upload()
df = pd.read_csv('data_sales.csv', header='infer')
print(df.shape)
df.head(5)
```

```python
# 1) Añadir nueva variable 'Amount' y mostrar las 5 primeras observaciones
df['Amount'] = df['Unit Price'] * df['Units']
print(f"1) Las 5 primeras observaciones con la nueva variable 'Amount':\n{df.head(5)}")
```

# 1) df.groupby('Region')
groupby('Region') agrupa las filas del DataFrame df en grupos basados en los valores de la columna Region.
Cada grupo representa todas las filas que tienen el mismo valor en la columna Region.

Por ejemplo, si la columna Region tiene valores como:

Region	Unit Price

Norte	10.5

Norte	12.0

Sur	8.0

Norte	11.0

Sur	9.5


El groupby generaría dos grupos:

Norte: Filas con Unit Price [10.5, 12.0, 11.0].

Sur: Filas con Unit Price [8.0, 9.5].

# 2) ['Unit Price']
Después de agrupar por Region, seleccionamos únicamente la columna Unit Price
para realizar cálculos sobre los valores de esta columna en cada grupo.

Por ejemplo:

Para el grupo Norte, seleccionamos los valores [10.5, 12.0, 11.0].

Para el grupo Sur, seleccionamos los valores [8.0, 9.5].


# 3) .mean()
Calcula el promedio (media) de los valores de la columna Unit Price para cada grupo.
Por ejemplo:

Norte: (10.5+12.0+11.0)/3=11.17

Sur: (8.0+9.5)/2=8.75

Resumen

Entrada: Un DataFrame con columnas Region y Unit Price.

Proceso: Agrupar por Region, seleccionar Unit Price y calcular el promedio por grupo.

Salida: Una tabla con el precio promedio de cada región.

```python
# 2) Precio unitario promedio de cada región utilizando groupby()
precio_promedio_regiones = df.groupby('Region')['Unit Price'].mean()
print(f"2) Precio unitario promedio de cada región:\n{precio_promedio_regiones}")
```

# 1. pivot_table
pivot_table es un método de pandas que permite reorganizar datos en forma tabular y resumirlos mediante funciones de agregación.
Es una forma flexible de agrupar y calcular estadísticas en datos. En este caso, está creando una tabla donde:

Las filas estarán indexadas por Region.
Los valores de la columna Unit Price se agregan utilizando la media (np.mean).


#Diferencia con groupby
Aunque el resultado es similar al de groupby en este caso, pivot_table ofrece más flexibilidad, como:

Trabajar con múltiples columnas en filas y columnas.
Especificar funciones de agregación adicionales.
Por ejemplo, podrías agregar más valores o funciones fácilmente.



```python
# 3) Precio unitario promedio de cada región utilizando pivot_table()
# Vertically
precio_promedio_vertical = df.pivot_table(index='Region', values='Unit Price', aggfunc=np.mean)
# Horizontally
precio_promedio_horizontal = df.pivot_table(columns='Region', values='Unit Price', aggfunc=np.mean)

print(f"3) Precio unitario promedio de cada región (vertical):\n{precio_promedio_vertical}")
print(f"3) Precio unitario promedio de cada región (horizontal):\n{precio_promedio_horizontal}")
```

```python
# 4) Precio unitario promedio por región en una sola línea con groupby()
precio_promedio_regiones_linea = df.groupby('Region')[['Unit Price']].mean()
print(f"4) Precio unitario promedio por región en una línea con groupby():\n{precio_promedio_regiones_linea}")
```

```python
# 5) Precio unitario promedio y unidades por región en una sola línea con pivot_table()
precio_y_unidades_regiones = df.pivot_table(index='Region', values=['Unit Price', 'Units'], aggfunc=np.mean)
print(f"5) Precio unitario promedio y unidades por región en una línea con pivot_table():\n{precio_y_unidades_regiones}")
```

#fill_value=0
 Reemplaza los valores NaN (que surgen cuando no hay datos para una combinación específica de Region e Item) con 0.

```python
# 6) Total de unidades por región y tipo de artículo con pivot_table(), rellenando con 0
total_unidades = df.pivot_table(index='Region', columns='Item', values='Units', aggfunc=np.sum, fill_value=0)
print(f"6) Total de unidades por región y tipo de artículo:\n{total_unidades}")
```

```python
# 7) Importe total de ventas por región y tipo de artículo con pivot_table(), rellenando con 0
total_ventas = df.pivot_table(index='Region', columns='Item', values='Amount', aggfunc=np.sum, fill_value=0)
print(f"7) Importe total de ventas por región y tipo de artículo:\n{total_ventas}")
```

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline

from google.colab import files
uploaded = files.upload()
df = pd.read_csv('data_coffeeshop.csv', header='infer', na_values=[' '])
print(df.shape)
df.head(5)
```

```python
# 1) Tabla de frecuencias para la variable 'Year of Start'
table = df['Year of Start'].value_counts()
table = table.sort_index()  # Ordena por año

# Visualización de la gráfica de línea
plt.plot(table.index, table.values, color='blue', alpha=0.5, linewidth=2, linestyle='--')
plt.xlim([1997, 2014])  # Limitar el rango de años
plt.xlabel('Year of Start')
plt.ylabel('Frequency')
plt.title('Frequency of Year of Start (1997-2014)')
plt.show()
```

```python
# 2) Divide los datos por el estado actual del negocio
df_in = df[df['Current State'] == 'In']  # En negocio
df_out = df[df['Current State'] == 'Out']  # Fuera de negocio

# Tablas de frecuencia para 'In' y 'Out'
table_in = df_in['Year of Start'].value_counts()
table_in = table_in.sort_index()  # Ordena por año
table_out = df_out['Year of Start'].value_counts()
table_out = table_out.sort_index()  # Ordena por año

# Visualización como dos gráficas de líneas superpuestas
fig = plt.figure(figsize=[8, 5])
my_axes = fig.add_axes([0, 0, 1, 1])  # Crear un solo eje

# Dibuja las dos líneas
my_axes.plot(table_in.index, table_in.values, color='blue', linestyle='--', label='In Business')
my_axes.plot(table_out.index, table_out.values, color='red', linestyle='-.', label='Out of Business')

# Añadir leyenda y limitar el rango de años
my_axes.legend(loc=0)
my_axes.set_xlim([1997, 2014])  # Limitar el rango de años
my_axes.set_xlabel('Year of Start')
my_axes.set_ylabel('Frequency')
my_axes.set_title('Business Status and Year of Start (1997-2014)')

plt.show()
```

A partir de los análisis, se observa una tendencia interesante:

Tendencia general:
Alrededor del año 2008, se empieza a notar un aumento pronunciado en el número de nuevos negocios (cafeterías, por ejemplo).
Desde 2008 hasta 2014, aunque el número de nuevos negocios crece, los negocios "In" (en funcionamiento) son más prevalentes que aquellos que han cerrado, lo que sugiere que la apertura de nuevas tiendas no ha sido completamente contrarrestada por el cierre de otras.

Descripción final:
Aunque hubo un auge de nuevas aperturas en los últimos años (2008-2014), la mayoría de los negocios que sobrevivieron están "In" en 2014, lo que indica una tendencia más positiva hacia la estabilidad de los negocios abiertos durante esos años.
